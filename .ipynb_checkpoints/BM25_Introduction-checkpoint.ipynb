    We will be using a trained pipeline en_core_web_sm is a small English pipeline trained on written web text (blogs, news, comments), that includes vocabulary, syntax and entities.YOu can find teh complete details of the trained pipeline can be found in (https://spacy.io/models/en). The struucture of teh pipeline looks as below : 

    We will be disabling the tagger , parser and ner component as we do not need them as per our task.

    Let us run for first 4 documents and see the internal outputs then we will just run it on the complete corpus efficient;y.

```python
# Preprocessing and Tokenizing 

import spacy

#load the trained model
nlp = spacy.load("en_core_web_sm")

# container for our tokenised corpus
tok_text = [] 

#lets just take the text column values in lower case 
series_of_docs = df.text.str.lower().values

#Tokenising using SpaCy:

no_of_doc = 0
for doc in nlp.pipe(series_of_docs, disable = ["tagger", "parser","ner"]):
    tok = [t.text for t in doc if t.is_alpha]
    print(tok)
    print('\n')
    tok_text.append(tok)
    no_of_doc += 1
    if no_of_doc == 2:
        break
```

```python
#the final container contents :
print('\n',len(tok_text))
print('\n',len(tok_text[0]),len(tok_text[1]))
print('\n',tok_text)
```

    Now lets run it on the full corpus. As the corpus is huge, what we would do is we will use another python package which can show us the progress bar.  

```python
# Preprocessing and Tokenizing 

from tqdm import tqdm

tok_text = [] 

for doc in tqdm(nlp.pipe(df.text.str.lower().values, disable = ["tagger", "parser","ner"])):
    tok = [t.text for t in doc if t.is_alpha]
    tok_text.append(tok)
```

    We could have done the same thing using nltk to tokenize and clean the text 

```python
    For that we would first clean the documents using regex patterns.
```

```python
import re
import string
```

```python
documents = df['text']

#create an empty container to hold the cleaned documents 
list_of_cleanedDoc = []

#lets traverse the documents one by one and clean it 
for d in tqdm(documents):
    
    # Remove Unicode
    d = re.sub(r'[^\x00-\x7F]+', ' ', d)
    
    # Remove Mentions
    d = re.sub(r'@\w+', '', d)
    
    # Lowercase the document
    d = d.lower()
    
    # Remove punctuations
    d = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', d)
    
    # Lowercase the numbers
    d = re.sub(r'[0-9]', '', d)
    
    # Remove the doubled space
    d = re.sub(r'\s{2,}', ' ', d)
    
    #append the cleaned doc to the container list
    list_of_cleanedDoc.append(d)
    
```

```python
print(len(list_of_cleanedDoc))
```

    Now that we have cleaned ...the next task is to 
    (a) remove stop words 
    (b) stemming , 
    (c) lemmatization and then 
    (d) tokenize.

```python
#declaring container for the tokenised words for each documents as we did using Spacy 
tok_text_2 = []
```

```python
from nltk.corpus import stopwords

#create the set of stop words 
stop_words = set(stopwords.words('english'))
```

    Although we have removed the punchuation using string while cleaning teh doc , I will show another way of removing the punctuation in cases any left

```python
#creating the regex object for the punctuation keywords 
punctuation = re.compile(r'[-.?!,":;()|0-9]')
```

    Before looping through the document lets just import the word_tokeniser from nltk library. In the code we will be looping through the documents, removing punctuations , tokenizing them into words and then will keep only teh word which are not stop words. 

```python
from nltk.tokenize import word_tokenize
```

```python
#traverse through the documents one by one 
for text in tqdm(list_of_cleanedDoc):
    
    #now remove all the punctuations from the text 
    text = punctuation.sub('', text)
    
    #lower each words and then tokenize it 
    tokenized_words = word_tokenize(text.lower())
    
    #now we process each tokenized words and then we keep it only if it is not in the stop words set
    cleaned_tokenized_words = [w for w in tokenized_words if not w.lower() in stop_words]
    
    #append each list of tokenized words to the main list of text_tokenised
    tok_text_2.append(cleaned_tokenized_words)
```

    You must have realised it took some time right ... because SpaCy is production optimised but the above code is written by us with the help of some python packages from scratch and certainly "efficiency" is not our goal here.

    Lets now start the process of Lemmatization. We have multiple options in nltk but we will go with the most common one here which is WordNetLemmatizer

```python
from nltk.stem import WordNetLemmatizer
```

```python
#create a lemmatizer object 
lemmatizer_wordnet = WordNetLemmatizer()
```

    lets create a smaple set to see how it performs

```python
sample_set = [['ram','playing','be','done'],
              ['where','is','he','going'],
              ['booked','went','srijeet','testing']]
```

```python
# we will go through the each document text_list one by one present in tok_text_2
for i in range(len(sample_set)):
    
    text_list = sample_set[i]
    
    # now for each word of the document(present as tokens in text_list) lets lemmatize
    for j in range(len(sample_set[i])):
        
        if sample_set[i][j] != '':
            
            #add the lemmtized words to the document level container
            sample_set[i][j] = lemmatizer_wordnet.lemmatize(sample_set[i][j], pos='v')
```

```python
# we will go through the each document text_list one by one present in tok_text_2
for i in tqdm(range(len(tok_text_2))):
    
    text_list = tok_text_2[i]
    
    # now for each word of the document(present as tokens in text_list) lets lemmatize
    for j in range(len(tok_text_2[i])):
        
        if tok_text_2[i][j] != '':
            
            #add the lemmtized words to the document level container
            tok_text_2[i][j] = lemmatizer_wordnet.lemmatize(tok_text_2[i][j], pos='v')
```

### Part 2 

    So what is BM25? It stands for ‘Best match 25’ (the other 24 attempts were clearly not very successful). It was released in 1994 at the third Text Retrieval Conference, yes there really was a conference dedicated to text retrieval…

    It is probably best thought of as tf-idf ‘on steroids’, implementing two key refinements:

   **Term frequency saturation**. BM25 provides diminishing returns for the number of terms matched against documents. This is fairly intuitive, if you looking to search for a specific term which is very common in documents then there should become a point where the number of occurrences of this term become less useful to the search.

**Document length**. BM25 considers document length in the matching process. Again, this is intuitive; if a shorter article contains the same number of terms that match as a longer article, then the shorter article is likely to be more relevant.

    These refinements also introduce two hyper-parameters to adjust the impact of these items on the ranking function. ‘k’ to tune the impact of term saturation and ‘b’ to tune document length.

    Usage

    For this example we'll be using the BM25Okapi algorithm, but the others are used in pretty much the same way.

    Initalizing : 

    First thing to do is create an instance of the BM25 class, which reads in a corpus of text and does some indexing on it:

```python
corpus = [
    "Hello there good man!",
    "It is quite windy in London",
    "How is the weather today?"
]

tokenized_corpus = [doc.split(" ") for doc in corpus]

print(tokenized_corpus)
```

```python
from rank_bm25 import BM25Okapi
bm25 = BM25Okapi(tokenized_corpus)
```

    Note that this package doesn't do any text preprocessing. If you want to do things like lowercasing, stopword removal, stemming, etc, you need to do it yourself.

    The only requirements is that the class receives a list of lists of strings, which are the document tokens.



    Ranking of documents
    
    Now that we've created our document indexes, we can give it queries and see which documents are the most relevant:

```python
query = "windy London"
tokenized_query = query.split(" ")

doc_scores = bm25.get_scores(tokenized_query)

print(type(doc_scores))

print('\n',doc_scores)
```

    Good to note that we also need to tokenize our query, and apply the same preprocessing steps we did to the documents in order to have an apples-to-apples comparison

    Instead of getting the document scores, you can also just retrieve the best documents with

```python
bm25.get_top_n(tokenized_query, corpus, n=1)
```

    And to analyze the performance speed we will use the python package time.

```python
import time

t0 = time.time()
results = bm25.get_top_n(query_tok, df.text.values, n = 3)
t1 = time.time()
print(f'Searched 50,000 records in {round(t1-t0,3) } seconds \n')
for i in results:
  print('\n',i)
```
